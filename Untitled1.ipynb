example1


```python
# define class
class Animal:
    def __init__(self, name):
        self.name = name

    def evolve(self):
        print(f"{self.name} is evolving!")
# subclass
class Mammal(Animal):
    def evolve(self):
        print(f"{self.name} evolved into a higher mammalian form.")
# subclass
class Bird(Animal):
    def evolve(self):
        print(f"{self.name} evolved into a flight-capable bird.")

# give values
human = Mammal("Human")
human.evolve()

sparrow = Bird("Sparrow")
sparrow.evolve()
```

example2

```python
# make a class
class CelesBody:
    def __init__(self, name, mass):
        self.name = name
        self.mass = mass
# subclass
class Planet(CelesBody):
    def __init__(self, name, mass, has_life):
        super().__init__(name, mass)
        self.has_life = has_life
# subclass
class Moon(CelesBody):
    def __init__(self, name, mass, planet):
        super().__init__(name, mass)
        self.planet = planet


earth = Planet("Earth", 5.97e24, True)
moon = Moon("Moon", 7.35e22, earth)
print(f"{moon.name} orbits around {moon.planet.name}")
```

example1

```python
class HackTool:
    def execute(self):
        pass

class Keylogger(HackTool):
    def execute(self):
        print("Recording keystrokes...")

class Exploit(HackTool):
    def execute(self):
        print("Exploiting system vulnerability...")

# Usage
tools = [Keylogger(), Exploit()]
for tool in tools:
    tool.execute()
```

example2

```python
class Organism:
    def adapt(self):
        print("Generic adaptation")

class Reptile(Organism):
    def adapt(self):
        print("Reptile is shedding skin to adapt to its environment.")

class Amphibian(Organism):
    def adapt(self):
        print("Amphibian is adapting to both land and water.")

# give value
organisms = [Reptile(), Amphibian()]
for org in organisms:
    org.adapt()
```

encapsulation

example1

```python
class Spacecraft:
    def __init__(self, name):
        self.name = name
        self.__fuel_level = 100  # Private attribute

    def __refuel(self):  # Private method
        self.__fuel_level = 100
        print(f"{self.name} refueled.")

    def launch(self):
        if self.__fuel_level > 50:
            print(f"{self.name} is launching!")
        else:
            print(f"{self.name} needs refueling.")
            self.__refuel()


apollo = Spacecraft("Apollo 11")
apollo.launch()
```

example2

```python
class HackingNet:
    def __init__(self, name):
        self.name = name
        self.__encryption_key = "supersecret"  # Private attribute

    def __decrypt_message(self, message):
        print(f"Decrypting message with key '{self.__encryption_key}': {message}")

    def receive_message(self, message):
        print(f"Message received: {message}")
        self.__decrypt_message(message)

# Usage
network = HackingNet("DarkNet")
network.receive_message("Encrypted payload")
```
