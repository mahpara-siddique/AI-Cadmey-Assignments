**Functions using 2 parameters**

```python
# example1
def add_numbers(a, b):
    result = a + b
    print(f"The sum of {a} and {b} is {result}.")
    return result
add_numbers(5, 3)
# it will add 2 numbers and show the result
```

```python
# example2
def multiply_numbers(a, b):
    result = a * b
    print(f"The product of {a} and {b} is {result}.")
    return result
multiply_numbers(4, 6)
# it will multiply 2 numbers and return the result
```

```python
# example3
def concatenate_strings(s1, s2):
    result = s1 + " " + s2
    print(f"The concatenated string is: '{result}'")
    return result
concatenate_strings("Black", "Python")
```

```python
# example4
def larger_number(a, b):
    larger = max(a, b)
    print(f"The larger number between {a} and {b} is {larger}.")
    return larger
larger_number(10, 20)
# it will print the larger number between 10 and 20
```

```python
# example5
def is_divisible(a, b):
    if b == 0:
        print("Division by zero is not allowed.")
        return False
    divisible = a % b == 0
    if divisible:
        print(f"{a} is divisible by {b}.")
    else:
        print(f"{a} is not divisible by {b}.")
    return divisible
is_divisible(7, 3)
# it will check if the number is divisible bt other or not and then return the result
```

**Functio using default values**

```python
# example1
def greet_user(name="Mahpara"):
    print(f"Hello, {name}!")
greet_user(name="Mahpara")
# here the program will greet the user with given name
```

```python
# example2
def rectangle_area(length, width):
    return length * width
rectangle_area(length=5, width=5)
# it will calculate the area of rectangle
```

```python
# example3
def custom_message(message, ending):
    print(message + ending)
custom_message(message="Thank you for reaching us.", ending="!")
```

```python
# example4
def power(base, exponent):
    return base ** exponent
power(base=3, exponent=2)
# it will print the 3 with the power of 2
```

```python
# example5
def countdown(start):
    for i in range(start, 0, -1):
        print(i)
    print("Time's up!")
countdown(start=5)
# it will display the countdown
```

**Functions taking input from user**

```python
# example1
def calculate_sum():
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))
    print(f"The sum is {num1 + num2}.")
calculate_sum()
# it will take input from user and then calculate the sum
```

```python
# example2
def check_even_odd():
    num = int(input("Enter a number: "))
    if num % 2 == 0:
        print(f"{num} is even.")
    else:
        print(f"{num} is odd.")
check_even_odd()
# it will check whether the user input number is even or odd
```

```python
# xample3
def string_length():
    user_input = input("Enter a string: ")
    print(f"The length of the string is {len(user_input)}.")
string_length()
# user will enter a string as input and program will tell the length of the strin g
```

```python
# example4
def is_palindrome():
    word = input("Enter a word: ")
    if word == word[::-1]:
        print(f"{word} is a palindrome.")
    else:
        print(f"{word} is not a palindrome.")
is_palindrome()
# it will display wheter the gibven input is palindrome or not
```

```python
# example5
def celsius_to_fahrenheit():
    celsius = float(input("Enter temperature in Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"{celsius}°C is equal to {fahrenheit}°F.")
celsius_to_fahrenheit()
# this program will convert the celsius temperature to farenheit
```
