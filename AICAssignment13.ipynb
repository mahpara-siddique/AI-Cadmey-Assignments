```python
# Library 1: Basic Math Operations
%%writefile library1.py
def add_numbers(a, b):
    return a + b

def subtract_numbers(a, b):
    return a - b
```

library2

```python
# Library 2: String Manipulations
%%writefile library2.py
def to_uppercase(s):
    return s.upper()

def reverse_string(s):
    return s[::-1]
```

library3

```python
# Library 3: List Operations
%%writefile library3.py
def find_max(lst):
    return max(lst)

def sort_list(lst):
    return sorted(lst)
```

library4

```python
# Library 4: File Handling
%%writefile library4.py
def write_to_file(filename, content):
    with open(filename, 'w') as file:
        file.write(content)
        return f"Content written to {filename}"

def read_from_file(filename):
    with open(filename, 'r') as file:
        return file.read()
```

library5

```python
# Library 5: Data Conversions
%%writefile library5.py
def int_to_binary(n):
    return bin(n)

def binary_to_int(b):
    return int(b, 2)
```

```python
# Main Script: Importing and Using Libraries
%%writefile main_script.py
import library1
import library2
import library3
import library4
import library5

# Call functions from library1
print("Add Numbers:", library1.add_numbers(10, 5))
print("Subtract Numbers:", library1.subtract_numbers(10, 5))

# Call functions from library2
print("To Uppercase:", library2.to_uppercase("hello"))
print("Reverse String:", library2.reverse_string("world"))

# Call functions from library3
print("Find Max:", library3.find_max([1, 2, 3, 4, 5]))
print("Sort List:", library3.sort_list([3, 1, 4, 5, 2]))

# Call functions from library4
filename = "testfile.txt"
print(library4.write_to_file(filename, "Hello, this is a test file."))
print("Read from File:", library4.read_from_file(filename))

# Call functions from library5
print("Integer to Binary:", library5.int_to_binary(10))
print("Binary to Integer:", library5.binary_to_int("1010"))
```
