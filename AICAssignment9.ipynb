**Classes in Python**

```python
# Example1
class MyClass: #defined a class
  def __init__(self, Value1 = None , Value2 = None):
    #initialize the class with default values
    self.object1 = Value1
    self.object2 = Value2
  def print_values(self):
    # print values of object1 and object2
    print(f"Object 1: {self.object1}")
    print(f"Object 2: {self.object2}")
order = MyClass()
# update the values of object
order.object1 = "Biology"
order.object2 = "Online Teacher"
# print the values
order.print_values()
```
