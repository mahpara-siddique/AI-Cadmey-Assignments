2 classes each with 2 subclasses

```python
# Define the first main class for the circulatory system
class Heart:
    def __init__(self, heart_rate):
        self.heart_rate = heart_rate  # Heart rate in beats per minute

    def display_heart_rate(self):
        return f"Heart Rate: {self.heart_rate} bpm"


# Subclass 1 of Heart
class Atrium(Heart):
    def __init__(self, heart_rate, blood_volume):
        super().__init__(heart_rate)
        self.blood_volume = blood_volume  # Blood volume in mL

    def display_atrium_info(self):
        return f"Atrium Blood Volume: {self.blood_volume} mL"


# Subclass 2 of Heart
class Ventricle(Heart):
    def __init__(self, heart_rate, pressure):
        super().__init__(heart_rate)
        self.pressure = pressure  # Pressure in mmHg

    def display_ventricle_info(self):
        return f"Ventricle Pressure: {self.pressure} mmHg"


# Define the second main class for the circulatory system
class Blood:
    def __init__(self, blood_type):
        self.blood_type = blood_type  # Blood type (A, B, AB, O)

    def display_blood_type(self):
        return f"Blood Type: {self.blood_type}"


# Subclass 1 of Blood
class RedBloodCells(Blood):
    def __init__(self, blood_type, hemoglobin_level):
        super().__init__(blood_type)
        self.hemoglobin_level = hemoglobin_level  # Hemoglobin level in g/dL

    def display_rbc_info(self):
        return f"Hemoglobin Level: {self.hemoglobin_level} g/dL"


# Subclass 2 of Blood
class WhiteBloodCells(Blood):
    def __init__(self, blood_type, wbc_count):
        super().__init__(blood_type)
        self.wbc_count = wbc_count  # White blood cell count (cells/Î¼L)

    def display_wbc_info(self):
        return f"WBC Count: {self.wbc_count} cells"


# Create instances of subclasses
# Heart
atrium = Atrium(heart_rate=72, blood_volume=70)
ventricle = Ventricle(heart_rate=72, pressure=120)

# Blood
rbc = RedBloodCells(blood_type="O+", hemoglobin_level=14.5)
wbc = WhiteBloodCells(blood_type="O+", wbc_count=6000)

# Print values
print(atrium.display_heart_rate())
print(atrium.display_atrium_info())

print(ventricle.display_heart_rate())
print(ventricle.display_ventricle_info())

print(rbc.display_blood_type())
print(rbc.display_rbc_info())

print(wbc.display_blood_type())
print(wbc.display_wbc_info())
```
