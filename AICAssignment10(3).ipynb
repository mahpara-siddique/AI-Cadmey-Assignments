**3 classes each with 3 subclasses**

```python
# class1
class Galaxy:
  # subclass1
    class Spiral:
        def __init__(self, name):
            self.name = name

        def display(self):
            print(f"Spiral Galaxy: {self.name}")
# subclass2
    class Elliptical:
        def __init__(self, name):
            self.name = name

        def display(self):
            print(f"Elliptical Galaxy: {self.name}")
# subclass3
    class Irregular:
        def __init__(self, name):
            self.name = name

        def display(self):
            print(f"Irregular Galaxy: {self.name}")

# class2
class Star:
  # subclass1
    class MainSequence:
        def __init__(self, name):
            self.name = name

        def display(self):
            print(f"Main Sequence Star: {self.name}")
# subclass2
    class RedGiant:
        def __init__(self, name):
            self.name = name

        def display(self):
            print(f"Red Giant Star: {self.name}")
# subclass3
    class WhiteDwarf:
        def __init__(self, name):
            self.name = name

        def display(self):
            print(f"White Dwarf Star: {self.name}")

# class3
class Planet:
  # subclass1
    class Terrestrial:
        def __init__(self, name):
            self.name = name

        def display(self):
            print(f"Terrestrial Planet: {self.name}")
# subclass2
    class GasGiant:
        def __init__(self, name):
            self.name = name

        def display(self):
            print(f"Gas Giant Planet: {self.name}")
# subclass3
    class IceGiant:
        def __init__(self, name):
            self.name = name

        def display(self):
            print(f"Ice Giant Planet: {self.name}")


# Galaxy instances
spiral = Galaxy.Spiral("Milky Way")
elliptical = Galaxy.Elliptical("M87")
irregular = Galaxy.Irregular("Large Magellanic Cloud")

spiral.display()
elliptical.display()
irregular.display()

# Star instances
main_sequence = Star.MainSequence("Sun")
red_giant = Star.RedGiant("Betelgeuse")
white_dwarf = Star.WhiteDwarf("Sirius B")

main_sequence.display()
red_giant.display()
white_dwarf.display()

# # Planet instances
terrestrial = Planet.Terrestrial("Earth")
gas_giant = Planet.GasGiant("Jupiter")
ice_giant = Planet.IceGiant("Neptune")

terrestrial.display()
gas_giant.display()
ice_giant.display()
```
