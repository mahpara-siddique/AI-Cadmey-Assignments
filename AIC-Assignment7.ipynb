**For Loop**

```python
# example1
for i in range(1, 6):
    print(i)
# by using range i print numbers here from 1 to 6 =(1,2,3,4,5)
```

```python
# example2
for i in range(0, 11, 2):
    print(i)
# here i print even numbers from 0 to 10 by stepping of 2
```

```python
# example3
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)
# the loop iterate over the list and print all the elements in the list on by one
```

```python
# example4
total = 0
for i in range(1, 101):
    total += i
print(total)
# added all numbers from 1 to 100 to calculate their sum
```

```python
# example5
word = "hello"
for char in word:
    print(char)
# it will print all the characters of the string
```

```python
# example6
numbers = [1, 2, 3, 1, 4, 1]
count = 0
for num in numbers:
    if num == 1:
        count += 1
print(count)
# it will tell how many time number 1 is there in list
```

```python
# eample7
squares = []
for i in range(1, 6):
    squares.append(i**2)
print(squares)
# it will print the list of squares of numbers from 1 to 5
```

```python
# example8
for i in range(1, 11):
    print(f"3 x {i} = {3 * i}")
# it will print the multiplication table of 3
```

```python
# example9
person = {'name': 'Alice', 'age': 25, 'city': 'New York'}
for key, value in person.items():
    print(f"{key}: {value}")
# it will print all key value pairs that are in dictionary
```

```python
# example10
for i in range(1,33):
    print(i)
    # it will print numers from 1 to 32
```

**Nested Loops**

```python
# example1
for i in range(1, 6):
    for j in range(1, 6):
        print(i * j, end=" ")
    print()
# it will print a 5x5 multiplication table from 1 to 5
```

```python
# example2
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()
# here's a * pattern that will be printed in right-angle triangle format
```

```python
# example3
numbers = [1, 2, 3]
total = 0
for x in numbers:
    for y in numbers:
        total += x * y
print(total)
# it will print the sum of each pair of numbers in list(1,2,3 2,4,6 3,6,9 =36)
```

```python
# example4
for i in range(1, 4):
    for j in range(1, 4):
        print(f"({i},{j})", end=" ")
    print()
# it will print coordinats for 3x3 grid
```

```python
# example5
colors = ['red', 'green', 'blue']
sizes = ['S', 'M', 'L']
for color in colors:
    for size in sizes:
        print(f"{color} - {size}")
# it will print all possible combinations of element from given lists
```

```python
# example6
for i in range(1, 6):
    for j in range(1, 6):
        print(i, end=" ")
    print()
# it will print a pattern from 1 to 5 in which each line contain same number
```

```python
# example7
for i in range(1, 10):
    for j in range(1, 10):
        print(f"{i * j:2}", end=" ")
    print()
it will print a multiplication table in 9x9 grid
```

```python
# exxample8
numbers = [1, 2, 3, 7, 8]
for i in numbers:
    for j in numbers:
        if i + j == 10:
            print(f"Pair: {i}, {j}")
# it will print the pair of numbers from the list that have sum equal to 10
```

```python
# example9
numbers = [1, 2, 3, 1, 4, 2]
count = 0
for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] == numbers[j]:
            count += 1
print(count)
# it will print , how many pairs of equal numbers are exist in the given list
```

```python
# example10
for i in range(3):
    for j in range(3):
        print(1, end=" ")
    print()
# it will print a 3x3 matrix where every element is 1
```
